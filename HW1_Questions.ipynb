# Question 2: Maximum Likelihood Estimate (20 pts)

100 students at UT were asked for their preference between studying or watching movies during lockdown. They reported their preferences by entering 0 or 1 where 0 represents studying and 1 represents movies. A random sample of 10 UT students yielded the following preferences:

0 1 1 0 1 0 0 1 1 1

A statistician suggests that these choices are assumed to arise by independent and identically distributed (i.i.d.) sampling from the following distribution and if the unknown parameter $q$ can be estimated, then she can provide more insights about the students' mental state during lockdown.

$$
\begin{equation}
\nonumber P(x) = \left\{
\begin{array}{l l}
    q& \quad \text{for  } x=1\\
1-q & \quad \text{for } x=0
\end{array} \right.
\end{equation}
$$

Based on the definitions given above, identify the likelihood function and derive the maximum likelihood estimator of $q$. Using the given sample, find a maximum likelihood estimate of $q$ as well.



# Answer 2

L(q) = P(Xn|q) = (10 C 6) (q)^6 (1-q)^4

l(q) = log(L(q)) = 6 log q + 4 log (1-q)

Using Derivative to find maximum?

dl/dq = 210 * 6/q - 210 * 4/(1-q) = 0

6/q - 4/(1-q) = 0

6/q = 4/(1-q)

6-6q = 4q

6 = 10q

q = .6 <- MLE Estimator

10C6 (0.6)^6 (0.4)4 = 0.25 <- MLE estimate

Use the following code to import the dataset to predict the housing price based on different variables. The categorical variables are removed and rows with missing variables have been pre-processed to remove NaN and infinity values. The MEDV(Median Value of owner-occupied homes) is the ** test label ** that needs to be predicted.

Variables: 

 * CRIM     : per capita crime rate by town
 * ZN       : proportion of residential land zoned for lots over 25,000 sq.ft.
 * INDUS    : proportion of non-retail business acres per town
 * NOX      : nitric oxides concentration (parts per 10 million)
 * RM       : average number of rooms per dwelling
 * AGE      : proportion of owner-occupied units built prior to 1940
 * DIS      : weighted distances to five Boston employment centres
 * RAD      : index of accessibility to radial highways
 * TAX      : full-value property-tax rate per 10,000 dollars
 * PTRATIO  : pupil-teacher ratio by town
 * LSTAT    : lower status of the population
 * MEDV     : Median value of owner-occupied homes in $1000's




```python
from sklearn import linear_model
from sklearn.metrics import mean_absolute_error,mean_squared_error
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
pd.options.mode.chained_assignment = None


df = pd.read_csv('data.csv')
df = df.replace([np.inf, -np.inf], np.nan)
df = df.fillna(0)

X = df.drop(['MEDV','CHAS','B'],axis=1)
y = df['MEDV']
```

# part b

```python
reg = linear_model.LinearRegression()

reg.fit(X, y) # seems to imply that MEDV is the target
print("Regression Coefficients:", reg.coef_, "\n") #w1, w2, ... wn
print("Regression Intercept:", reg.intercept_,"\n") #w0, may not be necessary
y_pred = reg.predict(X)
print("MSE:", mean_squared_error(y, y_pred))
```

# part c

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=50)

reg2 = linear_model.LinearRegression()
reg2.fit(X_train, y_train)
print("Regression Coefficients:", reg2.coef_, "\n") #w1, w2, ... wn
print("Regression Intercept:", reg2.intercept_,"\n") #w0, may not be necessary

y_pred_train = reg2.predict(X_train)
print("MSE using train:", mean_squared_error(y_train, y_pred_train))
print('Coefficient of determination using train:', r2_score(y_train, y_pred_train))

y_pred_test = reg2.predict(X_test)
print("MSE using test:", mean_squared_error(y_test, y_pred_test))
print('Coefficient of determination using test:', r2_score(y_test, y_pred_test))
```

## Solution - Bonus Qn



```python
# went ahead and did this bc it heavily builds from 5 lol

# want: NOX, PTRATIO, LSTAT, CRIM, INDUS, RM, RAD
# have: CRIM, ZN, INDUS, NOX, RM, AGE, DIS, RAD, TAX, PTRATIO,LSTAT  
# need to remove:  ZN, AGE, DIS, TAX
n = 506
p = 7
X_limited = X.drop(['ZN','AGE','TAX', 'DIS'],axis=1)
X_train_2, X_test_2, y_train_2, y_test_2 = train_test_split(X_limited, y, test_size=0.25, random_state=50)

reg3 = linear_model.LinearRegression()
reg3.fit(X_train_2, y_train_2)
print("PART A")
print("Regression Coefficients:", reg3.coef_, "\n") #w1, w2, ... wn
print("Regression Intercept:", reg3.intercept_,"\n") #w0, may not be necessary

y_pred_train_2 = reg3.predict(X_train_2)
print("MSE using train:", mean_squared_error(y_train_2, y_pred_train_2))
r2 = r2_score(y_train_2, y_pred_train_2)
print('Coefficient of determination using train:', r2)
r2_adj = 1-(1.0-r2)*(n-1.0)/(n-p-1.0)
print('Adjusted Coefficient of determination using train:', r2_adj, '\n')

y_pred_test_2 = reg3.predict(X_test_2)
print("MSE using test:", mean_squared_error(y_test_2, y_pred_test_2))
r2 = r2_score(y_test_2, y_pred_test_2)
print('Coefficient of determination using test:', r2)
r2_adj = 1-(1.0-r2)*(n-1.0)/(n-p-1.0)
print('Adjusted Coefficient of determination using test:', r2_adj, '\n')
```
